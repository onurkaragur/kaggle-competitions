# Load Kaggle Credentials

Load credentials from the root .env file.

```python
from dotenv import load_dotenv
import os

# Load environment variables from root .env file
load_dotenv('../.env')

# Set Kaggle credentials
os.environ['KAGGLE_USERNAME'] = os.getenv('KAGGLE_USERNAME')
os.environ['KAGGLE_KEY'] = os.getenv('KAGGLE_KEY')

print("Kaggle credentials loaded from root .env file.")
```

# Authenticate with Kaggle API

```python
import kaggle

kaggle.api.authenticate()

print("Authenticated with Kaggle API.")
```

# Download Competition Data

Set the competition name and download the data.

```python
# Replace 'your-competition-name' with the actual competition name (e.g., 'titanic', 'house-prices-advanced-regression-techniques')
competition_name = 'your-competition-name'

# Download the competition data
kaggle.api.competition_download_files(competition_name, path='.', quiet=False)

print(f"Competition data for '{competition_name}' downloaded successfully.")
```
