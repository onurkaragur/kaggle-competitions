# Load Kaggle Credentials

Load credentials from the root .env file.

```python
from dotenv import load_dotenv
import os

# Load environment variables from root .env file
load_dotenv('../.env')

# Set Kaggle credentials
os.environ['KAGGLE_USERNAME'] = os.getenv('KAGGLE_USERNAME')
os.environ['KAGGLE_KEY'] = os.getenv('KAGGLE_KEY')

print("Kaggle credentials loaded from root .env file.")
```

# Authenticate with Kaggle API

```python
import kaggle

kaggle.api.authenticate()

print("Authenticated with Kaggle API.")
```

```python
import os

# Create kaggle.json file
os.makedirs(os.path.expanduser('~/.kaggle'), exist_ok=True)
with open(os.path.expanduser('~/.kaggle/kaggle.json'), 'w') as f:
    f.write(f'{{"username": "{os.environ["KAGGLE_USERNAME"]}", "key": "{os.environ["KAGGLE_KEY"]}"}}')

print("kaggle.json created.")
```

# Download Competition Data

**Important:** Make sure you have joined the competition and accepted the rules on the Kaggle website before running this cell.

Set the competition name and download the data.

```python
# Download the competition data
import subprocess

competition_name = 'llm-classification-finetuning'  # Replace with actual competition name

result = subprocess.run(['kaggle', 'competitions', 'download', '-c', competition_name, '-p', '.'], capture_output=True, text=True)
if result.returncode == 0:
    print(f"Competition data for '{competition_name}' downloaded successfully.")
else:
    print(f"Failed to download data: {result.stderr}")
    print("Make sure the competition exists and you have accepted the rules via the Kaggle website.")
```
